```python
from differentiable_t1d.dynamics.kanderian_mvp import load_meals
import pandas as pd
```

```python
df = load_meals()
```

```python
df = df.dropna()
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()

ax.scatter(df[df.meal=="b"]["carbs"], df[df.meal=="b"]["tau"], label="Breakfast")
ax.scatter(df[df.meal=="l"]["carbs"], df[df.meal=="l"]["tau"], label="Lunch")
ax.scatter(df[df.meal=="d"]["carbs"], df[df.meal=="d"]["tau"], label="Dinner")
ax.scatter(df[df.meal=="s"]["carbs"], df[df.meal=="s"]["tau"], label="Snack")
ax.axhline(df["tau"].mean(), label="Mean", color="black")
ax.axhline(df["tau"].median(), label="Median", color="red")
ax.set(xlabel="carbs(g)", ylabel="tau_m(1/min)")
ax.legend()
```
